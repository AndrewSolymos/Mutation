```python
PATTERN_QUOTES = re.compile(
    r'(?:'
    r'"([^"\(\)]+)\(([^"\(\)]+)\)"?'   # left quote exists
    r'|'
    r'([^"\(\)]+)\(([^"\(\)]+)\)"'     # right quote exists
    r')'
)

# Regex:
# In column 2 only:
# Add a space after G/F/U/X if followed by lowercase letter
#
PATTERN_SPACING = re.compile(r'([GFUX])([a-z])')


# Regex:
# Replace hyphen with underscore
# EXCEPT when hyphen is part of ->
#
# Matches "-" not preceded by "-" and not followed by ">"
#
PATTERN_HYPHEN = re.compile(r'-(?!>)')


def transform_quotes(value: str) -> str:
    """
    Replace:

        "text1(text2)"
        "text1(text2
        text1(text2)"
        "text1(text2,text3)"

    with:

        text1_text2
        text1_text2_text3

    BUT do not touch:

        text1(text2)
    """

    def replacer(match):

        # Depending on which side had the quote,
        # one pair of groups will be None
        text1 = match.group(1) or match.group(3)
        inner = match.group(2) or match.group(4)

        text1 = text1.strip()
        inner = inner.strip()

        # Split comma-separated items
        parts = [p.strip() for p in inner.split(",")]

        # Join using underscores
        return "_".join([text1] + parts)

    return PATTERN_QUOTES.sub(replacer, value)


def add_spacing(value: str) -> str:
    """
    Add space after G/F/U/X when followed by lowercase letter.
    """

    return PATTERN_SPACING.sub(r'\1 \2', value)


def replace_hyphens(value: str) -> str:
    """
    Replace hyphens with underscores,
    except when part of the implies symbol '->'
    """

    return PATTERN_HYPHEN.sub("_", value)

combined_df['Spot LTL'] = combined_df['Spot LTL'].apply(transform_quotes).apply(add_spacing).apply(replace_hyphens)
combined_df['APs'] = combined_df['APs'].apply(transform_quotes).apply(add_spacing).apply(replace_hyphens)
```

```python

def make_balanced_batches(
    df,
    type_col="normalized_LTL_formula",
    source_col="source",
    n_batches=9,
    per_batch=8,
    random_state=42,
):
    rng = np.random.default_rng(random_state)

    batches = {i: [] for i in range(n_batches)}
    batch_sizes = np.zeros(n_batches, dtype=int)

    grouped = list(df.groupby([type_col, source_col]))
    rng.shuffle(grouped)

    for pair, group in grouped:

        idx = group.index.to_numpy().copy()
        rng.shuffle(idx)

        total_needed = n_batches * per_batch

        # --------------------------------------------------------------
        # enough rows -> exactly 10 per batch
        # --------------------------------------------------------------
        if len(idx) >= total_needed:

            selected = idx[:total_needed]

            for b in range(n_batches):
                rows = selected[b * per_batch:(b + 1) * per_batch]

                batches[b].extend(rows)
                batch_sizes[b] += len(rows)

        # --------------------------------------------------------------
        # fewer than 90 rows
        # distribute evenly with max difference = 1
        # --------------------------------------------------------------
        else:

            k = len(idx)

            base = k // n_batches
            remainder = k % n_batches

            # give extra rows to currently smallest batches
            ordered_batches = list(np.argsort(batch_sizes))

            counts = [base] * n_batches

            for i in range(remainder):
                counts[ordered_batches[i]] += 1

            ptr = 0

            for b in range(n_batches):

                take = counts[b]

                if take > 0:
                    rows = idx[ptr:ptr + take]

                    batches[b].extend(rows)
                    batch_sizes[b] += len(rows)

                    ptr += take

    batch_dfs = []

    for b in range(n_batches):

        batch_df = (
            df.loc[batches[b]]
            .sample(frac=1, random_state=random_state)
            .reset_index(drop=True)
        )

        batch_df["batch_id"] = b

        batch_dfs.append(batch_df)

    return batch_dfs

thisisit = make_balanced_batches(combined_df)


# Stack them vertically
thisisit = pd.concat(thisisit, ignore_index=True)
thisisit.sort_values(by=['batch_id']).reset_index(drop=True)
```

```python
import random
def sample_and_remove(data_dict, sample_size=10):
    samples = {}
    
    for source, formulas in data_dict.items():
        samples[source] = {}
        for formula, pairs_set in formulas.items():
            # Convert set to list to allow sampling
            pairs_list = list(pairs_set)
            
            # Determine how many we can actually take
            actual_sample_size = min(len(pairs_list), sample_size)
            
            # Randomly select items
            selected = random.sample(pairs_list, actual_sample_size)
            
            # Store them in our result
            samples[source][formula] = selected
            
            # Remove the selected items from the original dictionary
            # This ensures they won't be picked in the next "later time" call
            for item in selected:
                pairs_set.remove(item)
                
    return samples
first_batch = sample_and_remove(nested_dict, sample_size=12)
```
