```python
import pandas as pd
sumtotal=0
ltl_types = set()
import os
path2data = './processed_data/ConformalNL2LTL/'
data = os.listdir(path2data)
data = [x for x in data]
lista = []
for elem in data:
    df = pd.read_csv(path2data + '/' + elem, sep=';')
    df['source'] = 'ConformalNL2LTL'
    lista.append( df )
result = pd.concat(lista, ignore_index=True)
import re
def normalize_aps(formula):
    return re.sub(r'ap_\d+', 'prop', formula)

upper_limit=10


result['normalized_LTL_formula'] = result['lifted_LTL'].apply(normalize_aps)
result['normalized_LTL_formula'].value_counts().to_csv('dist/ConformalNL2LTL.csv')

sumtotal += result['normalized_LTL_formula'].value_counts().clip(upper=upper_limit).sum()
ltl_types |= set(result['normalized_LTL_formula'].value_counts().index.tolist())

nested_dict={}
nested_dict2 = {
    source: result.groupby('normalized_LTL_formula').apply(
        lambda x: set(zip(x['original NL'], x['Spot LTL'], x['APs']))
    ).to_dict()
    for source, result in result.groupby('source')
}

nested_dict = nested_dict2
```

```python
import random
def sample_and_remove(data_dict, sample_size=10):
    samples = {}
    
    for source, formulas in data_dict.items():
        samples[source] = {}
        for formula, pairs_set in formulas.items():
            # Convert set to list to allow sampling
            pairs_list = list(pairs_set)
            
            # Determine how many we can actually take
            actual_sample_size = min(len(pairs_list), sample_size)
            
            # Randomly select items
            selected = random.sample(pairs_list, actual_sample_size)
            
            # Store them in our result
            samples[source][formula] = selected
            
            # Remove the selected items from the original dictionary
            # This ensures they won't be picked in the next "later time" call
            for item in selected:
                pairs_set.remove(item)
                
    return samples
first_batch = sample_and_remove(nested_dict, sample_size=10)
```
